```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
plt.style.use("dark_background")
```

```python
N = 1000
d = np.random.normal(0, 1, N)
plt.hist(d)
```

```python
def is_sorted(d):
	return np.all(d[:-1] <= d[1:])

def ecdf(d, x):
	if not is_sorted(d): 
		d.sort()
	n = len(d)
	num_lte = (d <= x).sum()
	return num_lte/n
	

xs = np.linspace(-5, 5)
plt.plot(xs, [ecdf(d, x) for x in xs], label="ECDF")
X = scipy.stats.Normal()
plt.plot(xs, X.cdf(xs), label="theoretical CDF")
plt.legend()
```
