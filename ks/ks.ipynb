```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
plt.style.use("dark_background")
```

```python
np.random.seed(0) # reproduce
N = int(1e3)
a = np.random.normal(0, 1, N)
b = np.random.normal(1, 1, N)
opacity = 0.8
plt.hist(a, alpha=opacity)
plt.hist(b, alpha=opacity)
```

```python
def test(a, b, alpha=0.05):
	D, p = scipy.stats.ks_2samp(a, b)
	print(f"{D=}, {p=}")
	if p < alpha:
		print("Reject the null hypothesis: Samples come from different distributions.")
	else:
		print("Fail to reject the null hypothesis: Samples come from the same distribution.")
test(a,b)
```

```python
c = np.random.normal(0, 1, N)
plt.hist(a)
plt.hist(c)
```

```python
test(a,c)
```
